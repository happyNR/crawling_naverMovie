```python
from selenium import webdriver
import pandas as pd
import time
```

```python
driver = webdriver.Chrome('C:/Users/snari/Downloads/chromedriver_win32/chromedriver.exe')
```

```python
driver.get('https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=167635&target=after')
```

```python
num_text = []
review_text = []
result = {}
```

```python
for page in range(1,10):
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    
    num_list = soup.select('tr > td.ac.num')
    review_list = soup.select('tr > td.title')
    
    for i in range(len(review_list)):
        num_text.append(num_list[i].text)
        review_text.append(review_list[i].text.replace('말레피센트 2','').replace('신고','').strip())
        
    driver.find_element_by_css_selector('#old_content > div.paging > div > a.pg_next').click()
    time.sleep(0.5)
```

```python
result = {"number" : num_text, "review" : review_text}
df = pd.DataFrame(result)
```

```python
from tqdm import tqdm_notebook

num_text = []
review_text = []
result = {}


for page in tqdm_notebook(range(10)):
    base_url = "https://movie.naver.com/movie/point/af/list.nhn?st=mcode&sword=167635&target=after&page={}"
    url = base_url.format(page)
    html = requests.get(url)
    soup = BeautifulSoup(html.content, 'lxml')
    
    num_list = soup.select('tr > td.ac.num')
    review_list = soup.select('tr > td.title')
    
    for i in range(len(review_list)):
        num_text.append(num_list[i].text)
        review_text.append(review_list[i].text.replace('말레피센트 2','').replace('신고','').strip())
        
    time.sleep(0.5)
    
```

```python
result = {"number" : num_text, "review" : review_text}
df = pd.DataFrame(result)
```
